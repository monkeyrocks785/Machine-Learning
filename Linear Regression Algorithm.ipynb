```python
# importing modules and libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# defing arrays for data

a = 0
b = 900

x = np.append(x, (a + (b-a) * np.random.rand(1, 50)))
y = np.append(y, (a + (b-a) * np.random.rand(1, 50)))

print(f"Data : \nx = {x}\ny = {y}")
```

```python
# calculations

m_x = np.mean(x)
m_y = np.mean(y)
print(f"Mean : \nMean of x = {m_x}\nMean of y = {m_y}")

d_x = x - m_x
d_y = y - m_y
print(f"Deviation : \nDeviation in x = {d_x}\nDeviation in y = {d_y}")

sp_xy = np.sum(d_x * d_y)
print(f"Sum of product of deviations : {sp_xy}")

ss_xx = np.sum(d_x * d_x)
print(f"Sum of square of deviations in x : {ss_xx}")
```

```python
# slope and intercept calculations

m = sp_xy / ss_xx
print(f"Slope = {m}")

b = m_y - (m * m_x)
print(f"Intercept = {b}")
```

```python
# equation for prediction

y_pred = (m * x) + b

print(y_pred)
```

```python
# visualization

plt.scatter(x, y, color = 'r', s = 10)
plt.plot(x, y_pred)
```
